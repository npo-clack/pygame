````python
# プレイヤーの生成

#***************
# ライブラリのインポート
#***************
import pygame as pg                             # pg全体
from pygame.locals import *                     # 内部変数などにアクセスできる
import sys                                      # 

#***************
# 初期化
#***************
pg.init()                                       # 初期化
clock = pg.time.Clock()                         # 時間設定

#***************
# 画面設定
#***************
screen = pg.display.set_mode((800,600))         # 画面設定
pg.display.set_caption('shoot')                 # windownの左上のテキスト

#***************
# 登場人物の設定
#***************
# 操作キャラ
player = pg.Surface((50, 30))                   # 領域を確保
player.fill((0, 128, 255))                      # 青く塗りつぶす
player_rect = player.get_rect()                 # Rect(位置情報)の取得
player_rect.topleft=(400,540)                   # スタート位置を決める

#***************
# ゲーム内容
#***************
while True:

    #---------------
    # 画面の操作
    #---------------
    screen.fill((0, 0, 0))                   # 画面の色設定

    #---------------
    # 登場人物の操作
    #---------------
    
    # プレイヤーの操作
    keys = pg.key.get_pressed()
    if keys[pg.K_LEFT]:
        player_rect.move_ip(-5,0)
    if keys[pg.K_RIGHT]:
        player_rect.move_ip(5,0)

    #---------------
    # 画面の表示
    #---------------
    screen.blit(player,player_rect)             # 操作キャラを配置する
    pg.display.update()                         # 画面表示

    #---------------
    # 終了判定
    #---------------
    # バツボタンを押すとゲームが終了するという処理
    for event in pg.event.get():
        if event.type == QUIT:
            pg.quit()
            sys.exit()

    clock.tick(60)   
````

````python
# まっすぐ飛ぶ弾

#***************
# ライブラリのインポート
#***************
import pygame as pg                             # pg全体
from pygame.locals import *                     # 内部変数などにアクセスできる
import sys                                      # 

#***************
# 初期化
#***************
pg.init()                                       # 初期化
clock = pg.time.Clock()                         # 時間設定

#***************
# 画面設定
#***************
screen = pg.display.set_mode((800,600))         # 画面設定
pg.display.set_caption('shoot')                 # windownの左上のテキスト

#***************
# 登場人物の設定
#***************
# 操作キャラ
player = pg.Surface((50, 30))                   # 領域を確保
player.fill((0, 128, 255))                      # 青く塗りつぶす
player_rect = player.get_rect()                 # Rect(位置情報)の取得
player_rect.topleft=(400,540)                   # スタート位置を決める

# 弾の設定
player_bullets_info = {}
bullet_speed = 10

# 時刻の取得
current_time    = pg.time.get_ticks()           # 現在の時間を取得

#***************
# ゲーム内容
#***************
while True:
    current_time = pg.time.get_ticks()              # 現在の時間を取得
    #---------------
    # 画面の操作
    #---------------
    screen.fill((0, 0, 0))                   # 画面の色設定

    #---------------
    # 登場人物の操作
    #---------------
    
    # プレイヤーの操作
    keys = pg.key.get_pressed()
    if keys[pg.K_LEFT]:
        player_rect.move_ip(-5,0)
    if keys[pg.K_RIGHT]:
        player_rect.move_ip(5,0)

    # プレイヤーの弾の生成
    if pg.mouse.get_pressed()[0]:                       # マウスの左ボタンが押される条件

        # 弾の生成
        bullet = pg.Surface((5, 10))                    # 領域確保
        bullet.fill((255, 255, 255))                    # 白色に塗る
        bullet_rect = bullet.get_rect()                 # 位置情報を取得
        bullet_rect.midtop = player_rect.midtop         # 位置をプレイヤーの中心位置に設定

        player_bullets_info[current_time]={             # 操作キャラクターが打った弾の情報
            "surface":bullet,                           # 領域確保
            "rect":bullet_rect,                         # 位置情報
            "dx":0,                                     # 弾のX軸移動距離
            "dy":-bullet_speed                          # 弾のY軸移動距離
        }

    # プレイヤーの弾の移動
    for key in player_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        dx = player_bullets_info[key]["dx"]             # 弾のdxを取り出す
        dy = player_bullets_info[key]["dy"]             # 弾のdyを取り出す
        player_bullets_info[key]["rect"].move_ip(dx,dy) # 弾の位置を移動する


    #---------------
    # 画面の表示
    #---------------
    screen.blit(player,player_rect)             # 操作キャラを配置する

    # 弾を表示する
    for key in player_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        surface = player_bullets_info[key]["surface"]   # 弾のsurfaceを取り出す
        rect    = player_bullets_info[key]["rect"]      # 弾のrectを取り出す
        screen.blit(surface,rect)                       # 弾を表示する

    pg.display.update()                         # 画面表示

    #---------------
    # 終了判定
    #---------------
    # バツボタンを押すとゲームが終了するという処理
    for event in pg.event.get():
        if event.type == QUIT:
            pg.quit()
            sys.exit()

    clock.tick(60)   
````

````python
# 発射間隔

#***************
# ライブラリのインポート
#***************
import pygame as pg                             # pg全体
from pygame.locals import *                     # 内部変数などにアクセスできる
import sys                                      # 

#***************
# 初期化
#***************
pg.init()                                       # 初期化
clock = pg.time.Clock()                         # 時間設定

#***************
# 画面設定
#***************
screen = pg.display.set_mode((800,600))         # 画面設定
pg.display.set_caption('shoot')                 # windownの左上のテキスト

#***************
# 登場人物の設定
#***************
# 操作キャラ
player = pg.Surface((50, 30))                   # 領域を確保
player.fill((0, 128, 255))                      # 青く塗りつぶす
player_rect = player.get_rect()                 # Rect(位置情報)の取得
player_rect.topleft=(400,540)                   # スタート位置を決める

# 弾の設定
player_bullets_info = {}
bullet_speed = 10

# 時刻の取得
current_time    = pg.time.get_ticks()       # 現在の時間を取得
previous_time   = 0                         # 一つ前の弾が発射した時間

#***************
# ゲーム内容
#***************
while True:
    current_time = pg.time.get_ticks()              # 現在の時間を取得
    
    #---------------
    # 画面の操作
    #---------------
    screen.fill((0, 0, 0))                   # 画面の色設定

    #---------------
    # 登場人物の操作
    #---------------
    
    # プレイヤーの操作
    keys = pg.key.get_pressed()
    if keys[pg.K_LEFT]:
        player_rect.move_ip(-5,0)
    if keys[pg.K_RIGHT]:
        player_rect.move_ip(5,0)

    # プレイヤーの弾の生成
    if pg.mouse.get_pressed()[0] \
            and current_time-previous_time>100:         # マウスの左ボタンが押される条件+発射間隔
        bullet = pg.Surface((5, 10))                    # 領域確保
        bullet.fill((255, 255, 255))                    # 白色に塗る
        bullet_rect = bullet.get_rect()                 # 位置情報を取得
        bullet_rect.midtop = player_rect.midtop         # 位置をプレイヤーの中心位置に設定

        player_bullets_info[current_time]={             # 操作キャラクターが打った弾の情報
            "surface":bullet,                           # 領域確保
            "rect":bullet_rect,                         # 位置情報
            "dx":0,                                     # 弾のX軸移動距離
            "dy":-bullet_speed                          # 弾のY軸移動距離
        }
        previous_time = current_time                    # 一つ前の弾が発射した時間として保存


    # プレイヤーの弾の移動
    for key in player_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        dx = player_bullets_info[key]["dx"]             # 弾のdxを取り出す
        dy = player_bullets_info[key]["dy"]             # 弾のdyを取り出す
        player_bullets_info[key]["rect"].move_ip(dx,dy) # 弾の位置を移動する


    #---------------
    # 画面の表示
    #---------------
    screen.blit(player,player_rect)             # 操作キャラを配置する

    # 弾を表示する
    for key in player_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        surface = player_bullets_info[key]["surface"]   # 弾のsurfaceを取り出す
        rect    = player_bullets_info[key]["rect"]      # 弾のrectを取り出す
        screen.blit(surface,rect)                       # 弾を表示する

    pg.display.update()                         # 画面表示

    #---------------
    # 終了判定
    #---------------
    # バツボタンを押すとゲームが終了するという処理
    for event in pg.event.get():
        if event.type == QUIT:
            pg.quit()
            sys.exit()

    clock.tick(60)   
````

## マウスの方向に飛ぶ弾を作る
いよいよマウスの方向に飛ぶ弾をコーディングしていきます。  

ざっくり流れを説明すると、まずマウスの座標を取得します。これはメソッドで簡単に取得できます。その座標をつかって、弾の飛ぶ方向を計算します。  

弾の飛ぶ方向は、マウスのXY座標と操作キャラクターのXY座標の差をとれば良いです。けれど、この差をそのまま使ってしまうとダメなので更にひと工夫いれます。一つずつ見ていきましょう。

### マウスの座標を取得

|メソッド|説明|

### 弾が飛ぶ方向を計算
ここが少しややこしいです。 
  
マウスと操作キャラクターのXY座標の差をそのまま弾の移動量として扱うと、マウスが遠いほど弾が速くなってしまいます。今回はマウスの”方向”のみを取り出したいので、これではだめです。  

結論から言うと、aとbそれぞれをマウスと操作キャラクターとの距離cで割ります。その後に弾のスピードを掛けると、マウスの方向のみを取り出せます。弾のスピードは今回は10としてます。マウスと操作キャラクターとの距離cは三平方の定理で求まります。  

上記は一体何をしているのか？分からなければそのまま進んでもOKですが、解説しておきます。aとbそれぞれを距離cで割るとどうなるかというと、距離cが1のときの、aとbが求まります。そのあとに10を掛けているので、1フレームで斜めに10進むX方向の移動量とY方向の移動量が求まるわけですね。
