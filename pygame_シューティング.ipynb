````python
# 操作キャラクターの生成

#***************
# ライブラリのインポート
#***************
import pygame as pg                             # pg全体
from pygame.locals import *                     # 内部変数などにアクセスできる
import sys                                      # 

#***************
# 初期化
#***************
pg.init()                                       # 初期化
clock = pg.time.Clock()                         # 時間設定

#***************
# 画面設定
#***************
screen = pg.display.set_mode((800,600))         # 画面設定
pg.display.set_caption('shoot')                 # windownの左上のテキスト

#***************
# 登場人物の設定
#***************
# 操作キャラ
player = pg.Surface((50, 30))                   # 領域を確保
player.fill((0, 128, 255))                      # 青く塗りつぶす
player_rect = player.get_rect()                 # Rect(位置情報)の取得
player_rect.topleft=(400,540)                   # スタート位置を決める

#***************
# ゲーム内容
#***************
while True:

    #---------------
    # 画面の操作
    #---------------
    screen.fill((0, 0, 0))                   # 画面の色設定

    #---------------
    # 登場人物の操作
    #---------------
    
    # 操作キャラクターの操作
    keys = pg.key.get_pressed()
    if keys[pg.K_LEFT]:
        player_rect.move_ip(-5,0)
    if keys[pg.K_RIGHT]:
        player_rect.move_ip(5,0)

    #---------------
    # 画面の表示
    #---------------
    screen.blit(player,player_rect)             # 操作キャラを配置する
    pg.display.update()                         # 画面表示

    #---------------
    # 終了判定
    #---------------
    # バツボタンを押すとゲームが終了するという処理
    for event in pg.event.get():
        if event.type == QUIT:
            pg.quit()
            sys.exit()

    clock.tick(60)   
````

````python
# まっすぐ飛ぶ弾

#***************
# ライブラリのインポート
#***************
import pygame as pg                             # pg全体
from pygame.locals import *                     # 内部変数などにアクセスできる
import sys                                      # 

#***************
# 初期化
#***************
pg.init()                                       # 初期化
clock = pg.time.Clock()                         # 時間設定

#***************
# 画面設定
#***************
screen = pg.display.set_mode((800,600))         # 画面設定
pg.display.set_caption('shoot')                 # windownの左上のテキスト

#***************
# 登場人物の設定
#***************
# 操作キャラ
player = pg.Surface((50, 30))                   # 領域を確保
player.fill((0, 128, 255))                      # 青く塗りつぶす
player_rect = player.get_rect()                 # Rect(位置情報)の取得
player_rect.topleft=(400,540)                   # スタート位置を決める

# 弾の設定
player_bullets_info = {}
bullet_speed = 10

# 時刻の取得
current_time    = pg.time.get_ticks()           # 現在の時間を取得

#***************
# ゲーム内容
#***************
while True:
    current_time = pg.time.get_ticks()              # 現在の時間を取得
    #---------------
    # 画面の操作
    #---------------
    screen.fill((0, 0, 0))                   # 画面の色設定

    #---------------
    # 登場人物の操作
    #---------------
    
    # 操作キャラクターの操作
    keys = pg.key.get_pressed()
    if keys[pg.K_LEFT]:
        player_rect.move_ip(-5,0)
    if keys[pg.K_RIGHT]:
        player_rect.move_ip(5,0)

    # 操作キャラクターの弾の生成
    if pg.mouse.get_pressed()[0]:                       # マウスの左ボタンが押される条件

        # 弾の生成
        bullet = pg.Surface((5, 10))                    # 領域確保
        bullet.fill((255, 255, 255))                    # 白色に塗る
        bullet_rect = bullet.get_rect()                 # 位置情報を取得
        bullet_rect.midtop = player_rect.midtop         # 位置を操作キャラクターの中心位置に設定

        player_bullets_info[current_time]={             # 操作キャラクターが打った弾の情報
            "surface":bullet,                           # 領域確保
            "rect":bullet_rect,                         # 位置情報
            "dx":0,                                     # 弾のX軸移動距離
            "dy":-bullet_speed                          # 弾のY軸移動距離
        }

    # 操作キャラクターの弾の移動
    for key in player_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        dx = player_bullets_info[key]["dx"]             # 弾のdxを取り出す
        dy = player_bullets_info[key]["dy"]             # 弾のdyを取り出す
        player_bullets_info[key]["rect"].move_ip(dx,dy) # 弾の位置を移動する


    #---------------
    # 画面の表示
    #---------------
    screen.blit(player,player_rect)             # 操作キャラを配置する

    # 操作キャラクターの弾を表示する
    for key in player_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        surface = player_bullets_info[key]["surface"]   # 弾のsurfaceを取り出す
        rect    = player_bullets_info[key]["rect"]      # 弾のrectを取り出す
        screen.blit(surface,rect)                       # 弾を表示する

    pg.display.update()                         # 画面表示

    #---------------
    # 終了判定
    #---------------
    # バツボタンを押すとゲームが終了するという処理
    for event in pg.event.get():
        if event.type == QUIT:
            pg.quit()
            sys.exit()

    clock.tick(60)   
````

## マウスの方向に飛ぶ弾を作る
いよいよマウスの方向に飛ぶ弾をコーディングしていきます。  

ざっくり流れを説明すると、まずマウスの座標を取得します。これはメソッドで簡単に取得できます。その座標をつかって、弾の飛ぶ方向を計算します。  

弾の飛ぶ方向は、マウスのXY座標と操作キャラクターのXY座標の差をとれば良いです。けれど、この差をそのまま使ってしまうとダメなので更にひと工夫いれます。一つずつ見ていきましょう。

### マウスの座標を取得
マウスの座標を取得もメソッドで簡単に取得できます。

|メソッド|説明|
|-----|------|
|pg.mouse.get_pos() 　                |マウスのXY座標を取得する|


### 弾が飛ぶ方向を計算
ここが少しややこしいです。 
  
マウスと操作キャラクターのXY座標の差をそのまま弾の移動量として扱うと、マウスが遠いほど弾が速くなってしまいます。今回はマウスの”方向”のみを取り出したいので、これではだめです。  

結論から言うと、aとbそれぞれをマウスと操作キャラクターとの距離cで割ります。その後に弾のスピードを掛けると、マウスの方向のみを取り出せます。弾のスピードは今回は10としてます。マウスと操作キャラクターとの距離cは三平方の定理で求まります。  

上記は一体何をしているのか？分からなければそのまま進んでもOKですが、解説しておきます。aとbそれぞれを距離cで割るとどうなるかというと、距離cが1のときの、aとbが求まります。そのあとに10を掛けているので、1フレームで斜めに10進むX方向の移動量とY方向の移動量が求まるわけですね。これで操作キャラクターとマウスの距離に関係なく一定の速度で弾が発射できます。

````python
# マウスの方向に飛ぶ弾

#***************
# ライブラリのインポート
#***************
import pygame as pg                             # pg全体
from pygame.locals import *                     # 内部変数などにアクセスできる
import sys                                      # 
import math                                     # 計算用のライブラリ

#***************
# 初期化
#***************
pg.init()                                       # 初期化
clock = pg.time.Clock()                         # 時間設定

#***************
# 画面設定
#***************
screen = pg.display.set_mode((800,600))         # 画面設定
pg.display.set_caption('shoot')                 # windownの左上のテキスト

#***************
# 登場人物の設定
#***************
# 操作キャラ
player = pg.Surface((50, 30))                   # 領域を確保
player.fill((0, 128, 255))                      # 青く塗りつぶす
player_rect = player.get_rect()                 # Rect(位置情報)の取得
player_rect.topleft=(400,540)                   # スタート位置を決める

# 弾の設定
player_bullets_info = {}
bullet_speed = 10

# 時刻の取得
current_time    = pg.time.get_ticks()       # 現在の時間を取得
previous_time   = 0                         # 一つ前の弾が発射した時間

#***************
# ゲーム内容
#***************
while True:
    current_time = pg.time.get_ticks()              # 現在の時間を取得
    
    #---------------
    # 画面の操作
    #---------------
    screen.fill((0, 0, 0))                   # 画面の色設定

    #---------------
    # 登場人物の操作
    #---------------
    
    # 操作キャラクターの操作
    keys = pg.key.get_pressed()
    if keys[pg.K_LEFT]:
        player_rect.move_ip(-5,0)
    if keys[pg.K_RIGHT]:
        player_rect.move_ip(5,0)

    # 操作キャラクターの弾の生成
    if pg.mouse.get_pressed()[0] \
            and current_time-previous_time>100:         # マウスの左ボタンが押される条件+発射間隔
        
        # 弾の飛ぶ方向の計算
        mouse_x, mouse_y = pg.mouse.get_pos()           # マウスの座標を取得
        player_x,player_y = player_rect.midtop          # 操作キャラクターの座標を取得
        dx = mouse_x - player_x                         # マウスと操作キャラクターのX座標の差を計算
        dy = mouse_y - player_y                         # マウスと操作キャラクターのY座標の差を計算
        length = math.sqrt(dx**2+dy**2)                 # マウスと操作キャラクターの距離を計算    
        dx = (dx/length)*bullet_speed                   # マウスと操作キャラクターの距離を10にするためのX座標の差を計算
        dy = (dy/length)*bullet_speed                   # マウスと操作キャラクターの距離を10にするためのY座標の差を計算

        # 弾の生成
        bullet = pg.Surface((5, 10))                    # 領域確保
        bullet.fill((255, 255, 255))                    # 白色に塗る
        bullet_rect = bullet.get_rect()                 # 位置情報を取得
        bullet_rect.midtop = player_rect.midtop         # 位置を操作キャラクターの中心位置に設定

        player_bullets_info[current_time]={             # 操作キャラクターが打った弾の情報
            "surface":bullet,                           # 領域確保
            "rect":bullet_rect,                         # 位置情報
            "dx":dx,                                    # 弾のX軸移動距離
            "dy":dy                                     # 弾のY軸移動距離
        }
        previous_time = current_time                    # 一つ前の弾が発射した時間として保存


    # 操作キャラクターの弾の移動
    for key in player_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        dx = player_bullets_info[key]["dx"]             # 弾のdxを取り出す
        dy = player_bullets_info[key]["dy"]             # 弾のdyを取り出す
        player_bullets_info[key]["rect"].move_ip(dx,dy) # 弾の位置を移動する


    #---------------
    # 画面の表示
    #---------------
    screen.blit(player,player_rect)             # 操作キャラを配置する

    # 操作キャラクターの弾を表示する
    for key in player_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        surface = player_bullets_info[key]["surface"]   # 弾のsurfaceを取り出す
        rect    = player_bullets_info[key]["rect"]      # 弾のrectを取り出す
        screen.blit(surface,rect)                       # 弾を表示する

    pg.display.update()                         # 画面表示

    #---------------
    # 終了判定
    #---------------
    # バツボタンを押すとゲームが終了するという処理
    for event in pg.event.get():
        if event.type == QUIT:
            pg.quit()
            sys.exit()

    clock.tick(60)   
````

## 敵キャラクターと操作キャラクターの弾の衝突判定

忘れていたら確認しておいてください。  

今回の衝突判定は少しややこしいです。  
敵キャラクターも複数いて、操作キャラクターの弾も複数あるからですね。  
複数あるものを使いたいときは、forでループします。

まずは```enemy_info```からforで敵の情報を一つずつとりだして、さらに```player_bullets_info```
もforで情報を取り出します。二重のforループを使う訳ですね。それぞれの位置情報が取り出せれば、あとは```colliderect```で衝突を判定して、衝突していれば、```enemy_info```の中の敵の一つを```del```で消せばOKです。このとき衝突した```player_bullets_info```の一つも消しています。これをしないと弾が敵を貫通する動作になります。

ここで一つ注意点があります。   
forループ中に```enemy_info```と```player_bullets_info```の中身が消えると、
forループがバグってしまいます。今取り出している情報が途中で消えてしまうと、プログラム側が何を見ているのかがわからなくなってしまうからですね。これを防ぐために、```copy```をつかって、```enemy_info```と```player_bullets_info```のコピーでループしています。

````python
# 弾の衝突判定

#***************
# ライブラリのインポート
#***************
import pygame as pg                             # pg全体
from pygame.locals import *                     # 内部変数などにアクセスできる
import sys                                      # 
import math                                     # 計算用のライブラリ
import random                                   # ランダムな数の生成用ライブラリ

#***************
# 初期化
#***************
pg.init()                                       # 初期化
clock = pg.time.Clock()                         # 時間設定

#***************
# 画面設定
#***************
screen = pg.display.set_mode((800,600))         # 画面設定
pg.display.set_caption('shoot')                 # windownの左上のテキスト

# 背景設定
back = pg.image.load("./Asset/imag/シューティング/pipo-battlebg020.jpg") # 画像を読み込み
back_rect = back.get_rect()                     # 背景の位置情報

#***************
# 登場人物の設定
#***************
# 操作キャラ
hunes = pg.image.load("./Asset/imag/シューティング/pipo-airship01.png") # 画像読み込み
hune  = hunes.subsurface(0,96,32,32)          
hune = pg.transform.scale(hune,(50,50))       # 爆破の画像を圧縮

player = hune #pg.Surface((50, 30))                   # 領域を確保
player_rect = player.get_rect()                 # Rect(位置情報)の取得
player_rect.topleft=(400,540)                   # スタート位置を決める

# 敵キャラクターの設定
enemies = pg.image.load("./Asset/imag/シューティング/pipo-boss004.png") # 画像読み込み
enemy = pg.transform.scale(enemies,(120,60))       # 爆破の画像を圧縮
enemy_info = {}

# 弾の設定
player_bullets = pg.image.load("./Asset/imag/シューティング/pipo-hiteffect007.png") # 画像読み込み
player_bullet = pg.transform.scale(player_bullets,(20,20))       # 爆破の画像を圧縮

enemy_bullets = pg.image.load("./Asset/imag/シューティング/pipo-hiteffect011.png") # 画像読み込み
enemy_bullet = pg.transform.scale(enemy_bullets,(20,20))       # 爆破の画像を圧縮

enemy_info = {}
player_bullets_info = {}
enemy_bullets_info = {}
bullet_speed = 10

# エフェクト
bakuhas = pg.image.load("./Asset/imag/シューティング/pipo-btleffect003.png") # 爆破の画像を読み込み
bakuha = bakuhas.subsurface(0,0,120,120) 
bakuha.set_colorkey((255,255,255)) 
bakuha_info = {}

# 時刻の取得
current_time    = pg.time.get_ticks()       # 現在の時間を取得
previous_time   = 0                         # 一つ前の弾が発射した時間

# 操作キャラクターが弾に当たったフラグ
hit_flag = 0

#***************
# ゲーム内容
#***************
while True:
    current_time = pg.time.get_ticks()              # 現在の時間を取得
    
    #---------------
    # 画面の操作
    #---------------
    screen.fill((0, 0, 0))                   # 画面の色設定

    #---------------
    # 登場人物の操作
    #---------------
    
    # 操作キャラクターの操作
    keys = pg.key.get_pressed()
    if keys[pg.K_LEFT]:
        player_rect.move_ip(-5,0)
    if keys[pg.K_RIGHT]:
        player_rect.move_ip(5,0)

    # 操作キャラクターの弾の生成
    if pg.mouse.get_pressed()[0] \
            and current_time-previous_time>100:         # マウスの左ボタンが押される条件+発射間隔
        
        # 弾の飛ぶ方向の計算
        mouse_x, mouse_y = pg.mouse.get_pos()           # マウスの座標を取得
        player_x,player_y = player_rect.midtop          # 操作キャラクターの座標を取得
        dx = mouse_x - player_x                         # マウスと操作キャラクターのX座標の差を計算
        dy = mouse_y - player_y                         # マウスと操作キャラクターのY座標の差を計算
        length = math.sqrt(dx**2+dy**2)                 # マウスと操作キャラクターの距離を計算    
        dx = (dx/length)*bullet_speed                   # マウスと操作キャラクターの距離を10にするためのX座標の差を計算
        dy = (dy/length)*bullet_speed                   # マウスと操作キャラクターの距離を10にするためのY座標の差を計算

        # 弾の生成
        bullet = player_bullet#pg.Surface((5, 10))                    # 領域確保
        #bullet.fill((255, 255, 255))                    # 白色に塗る
        bullet_rect = bullet.get_rect()                 # 位置情報を取得
        bullet_rect.midtop = player_rect.midtop         # 位置を操作キャラクターの中心位置に設定

        player_bullets_info[current_time]={             # 操作キャラクターが打った弾の情報
            "surface":bullet,                           # 領域確保
            "rect":bullet_rect,                         # 位置情報
            "dx":dx,                                    # 弾のX軸移動距離
            "dy":dy                                     # 弾のY軸移動距離
        }
        previous_time = current_time                    # 一つ前の弾が発射した時間として保存


    # 操作キャラクターの弾の移動
    for key in player_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        dx = player_bullets_info[key]["dx"]             # 弾のdxを取り出す
        dy = player_bullets_info[key]["dy"]             # 弾のdyを取り出す
        player_bullets_info[key]["rect"].move_ip(dx,dy) # 弾の位置を移動する


    # 敵キャラクター生成
    if random.randint(1, 60) == 1:                      # 1/60の確率で敵キャラクターを生成
        enemy = enemy#pg.Surface((40, 30))                    # 領域確保
        #enemy.fill((255, 0, 0))                         # 赤色に塗る
        enemy_rect = enemy.get_rect()                   # 位置情報を取得
        enemy_x = random.randint(0, 800-20)             # 敵のX座標をランダムに決める
        enemy_rect.midtop=(enemy_x,0)                   # 敵の初期位置を決める
        enemy_info[current_time] = {
            "surface":enemy,                           # 領域確保
            "rect":enemy_rect,                         # 位置情報
            "dx":0,                                    # 敵のX軸移動距離
            "dy":2                                     # 敵のY軸移動距離
        }

    # 敵キャラクターの移動
    for key in enemy_info.keys():                       # 敵キャラクター情報でループ
        dx = enemy_info[key]["dx"]                      # dxを取り出す
        dy = enemy_info[key]["dy"]                      # dyを取り出す
        enemy_info[key]["rect"].move_ip(dx,dy)          # 位置を移動する


    # 敵キャラクターの弾の生成
    for key in enemy_info.keys():
        if random.randint(1, 20) == 1:                      # 1/30の確率で敵キャラクターを生成
            # 弾の生成
            bullet = enemy_bullet#pg.Surface((5, 10))                    # 領域確保
            #bullet.fill((255, 255, 255))                    # 白色に塗る
            bullet_rect = bullet.get_rect()                 # 位置情報を取得

            rect = enemy_info[key]['rect']                  # 敵キャラクターの位置情報を取得
            bullet_rect.midtop =  rect.midbottom            # 位置を操作キャラクターの中心位置に設定

            enemy_bullets_info[current_time]={              # 操作キャラクターが打った弾の情報
                "surface":bullet,                           # 領域確保
                "rect":bullet_rect,                         # 位置情報
                "dx":0,                                     # 弾のX軸移動距離
                "dy":5                                      # 弾のY軸移動距離
            }

    # 敵キャラクターの弾の移動
    for key in enemy_bullets_info.keys():               # 敵キャラクターの弾の情報でループ
        dx = enemy_bullets_info[key]["dx"]             # 弾のdxを取り出す
        dy = enemy_bullets_info[key]["dy"]             # 弾のdyを取り出す
        enemy_bullets_info[key]["rect"].move_ip(dx,dy) # 弾の位置を移動する


   # 敵キャラクターと操作キャラクターの弾の衝突
    enemy_info_copy =  enemy_info.copy()                    # 衝突判定のためにコピーを作成
    player_bullets_info_copy =  player_bullets_info.copy()  # 衝突判定のためにコピーを作成

    for key1 in enemy_info_copy.keys():                     # 敵キャラクターの弾の情報でループ
        for key2 in player_bullets_info_copy.keys():        # 操作キャラクターの弾の情報でループ
            enemy_rect = enemy_info_copy[key1]["rect"]      # 敵キャラクターの位置情報を取得
            bullet_rect = player_bullets_info_copy[key2]["rect"] # 操作キャラクターの弾の位置情報を取得
            if enemy_rect.colliderect(bullet_rect):         # 衝突判定
                bakuha_info[key2] = {
                    "rect"  : enemy_rect,
                    "cnt"   : 0,
                }       
                del enemy_info[key1]                        # 衝突した敵キャラクターを削除
                del player_bullets_info[key2]               # 衝突した弾を削除

    # 操作キャラクターと敵キャラクターの弾の衝突
    for key in enemy_bullets_info.keys():               # 敵キャラクターの弾の情報でループ
        bullet_rect = enemy_bullets_info[key]["rect"]   # 敵キャラクターの弾の位置情報を取得
        if player_rect.colliderect(bullet_rect):        # 衝突判定
            hit_flag = 1                                # 衝突フラグを立てる


    #---------------
    # 画面の表示
    #---------------
    screen.blit(back,back_rect)                  # 背景を配置する
    screen.blit(player,player_rect)             # 操作キャラを配置する

    # 操作キャラクターの弾を表示する
    for key in player_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        surface = player_bullets_info[key]["surface"]   # 弾のsurfaceを取り出す
        rect    = player_bullets_info[key]["rect"]      # 弾のrectを取り出す
        screen.blit(surface,rect)                       # 弾を表示する

    # 敵キャラクターを表示する
    for key in enemy_info.keys():                       # 敵キャラクター情報でループ
        surface = enemy_info[key]["surface"]            # surfaceを取り出す
        rect    = enemy_info[key]["rect"]               # rectを取り出す
        screen.blit(surface,rect)                       # 敵キャラクターを表示する

    # 操作キャラクターの弾を表示する
    for key in enemy_bullets_info.keys():              # 操作キャラクターの弾の情報でループ
        surface = enemy_bullets_info[key]["surface"]   # 弾のsurfaceを取り出す
        rect    = enemy_bullets_info[key]["rect"]      # 弾のrectを取り出す
        screen.blit(surface,rect)                       # 弾を表示する
    
    # 爆破エフェクトを表示する
    bakuha_info_copy =  bakuha_info.copy()              # 爆爆破判定のためにコピーを作成
    for key in bakuha_info_copy:
        cnt = bakuha_info[key]["cnt"]
        rect = bakuha_info[key]["rect"]

        if cnt // 5 == 3:
            del bakuha_info[key]

        bakuha = bakuhas.subsurface(120*(cnt//5),0,120,120) 
        bakuha.set_colorkey((255,255,255)) 
        #bakuha = pg.transform.scale(bakuha,(25,25))       # 爆破の画像を圧縮
        screen.blit(bakuha,rect)

        if cnt // 5 != 3:
            bakuha_info[key]["cnt"] += 1



    pg.display.update()                         # 画面表示
    #---------------
    # 終了判定
    #---------------
    # バツボタンを押すとゲームが終了するという処理
    for event in pg.event.get():
        if event.type == QUIT:
            pg.quit()
            sys.exit()
    
    # ゲームオーバー判定
    if hit_flag == 1:                           # 衝突フラグが立っていたら
        pg.quit()                               # 終了
        sys.exit()

    clock.tick(60)   
````
